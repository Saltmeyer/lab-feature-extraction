## Lab Feature Extraction

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
customer_df.columns = list(map(lambda x:x.lower(),customer_df.columns))
customer_df.columns = list(map(lambda x:x.replace(' ','_'),customer_df.columns))
```

```python
categoricals = customer_df.select_dtypes(np.object)
```

```python
categoricals.head(60)
```

```python
categoricals = categoricals.drop(['customer'], axis=1)
```

```python
# Function to plot all variables
def plot_all(df):
    for column in df:
        sns.displot(x=column, data=df)

    return None
```

```python
plot_all(categoricals)
```

```python
# Some variables contain categories with either very large or very low amount of values. 
# Some categories could also be merged if it makes sense in the context. This what I am going to do in the next steps.
```

```python
def clean_employment_col(x):
    if x in ['Medical Leave','Disabled','Retired']:
        return 'Inactive'
    else:
        return x
categoricals['employmentstatus'] = categoricals['employmentstatus'].apply(clean_employment_col)
# Merging Medical Leave, Disabled and Retired
```

```python
def clean_employment_col(x):
    if x in ['Medical Leave','Disabled','Retired']:
        return 'Inactive'
    else:
        return x
categoricals['employmentstatus'] = categoricals['employmentstatus'].apply(clean_employment_col)
# Merging Medical Leave, Disabled and Retired
```

```python
def clean_coverage_col(x):
    if x in ['Premium']:
        return 'Extended'
    else:
        return x
categoricals['coverage'] = categoricals['coverage'].apply(clean_coverage_col)
# Merging Premium into Extended
```

```python
def clean_education_col(x):
    if x in ['Master','Doctor']:
        return 'Postgraduate'
    else:
        return x
categoricals['education'] = categoricals['education'].apply(clean_education_col)
# Merging Doctor and Master
```

```python
def clean_vehicleclass_col(x):
    if x in ['Luxury Car', 'Sports Car','Luxury SUV']:
        return 'Luxury Car'
    else:
        return x
categoricals['vehicle_class'] = categoricals['vehicle_class'].apply(clean_vehicleclass_col)
# Merging all luxury categories
```

```python
def clean_policy_col(x):
    if x in ['Corporate L3', 'Personal L3','Special L3']:
        return 'L3'
    elif x in ['Corporate L2', 'Personal L2','Special L2']:
        return 'L2'
    else:
        return 'L1'
categoricals['policy'] = categoricals['policy'].apply(clean_policy_col)
# Merging Corporates Personal and Special
```

```python
# Date 
categoricals['effective_to_date'] = pd.to_datetime(categoricals['effective_to_date'])
```
